```python
import polars as pl
```

```python
WP497 = pl.read_csv(
    "../Data/Wikipathways/WP497-datanodes.tsv",
    separator="\t"
).filter(
    pl.col("Identifier").str.contains("eccode:")
).with_columns(
    pl.col("Identifier").str.extract(r"eccode:(\d+\.\d+\.\d+\.\d+)").alias("eccode")
)

pl.Config.set_tbl_rows(16)
display(WP497)
```

```python
eccodes = WP497.select(pl.col("eccode")).sort("eccode").unique().write_csv(
    "../Data/Wikipathways/WP497-eccodes.tsv",
    separator="\t",
    include_header=False
)
```

```python
# !bash ../scripts/togoid_convert.sh ../Data/Wikipathways/WP497-eccodes.csv ec,rhea,uniprot,taxonomy ../Data/Wikipathways/WP497-eccodes-converted.tsv
```

```python
togoid_convert1 = pl.read_csv(
    "../Data/Wikipathways/WP497-eccodes-converted.tsv",
    separator="\t"
)

uniprot_ids = togoid_convert1.select(pl.col("uniprot_id")).sort("uniprot_id").unique().write_csv(
    "../Data/Wikipathways/WP497-uniprot_ids.tsv",
    separator="\t",
    include_header=False
)

display(togoid_convert1)
```

```python
uniprot_taxonomy = pl.read_csv(
    "../Data/Wikipathways/WP497-uniprot_taxonomy_converted.tsv",
    separator="\t"
)

display(uniprot_taxonomy)
```

```python
togoid_convert_join = togoid_convert1.join(
    uniprot_taxonomy,
    on="uniprot_id",
    how="inner"
).filter(
    (pl.col("taxonomy_id") == 9606) |
    (pl.col("taxonomy_id") == 7227) |
    (pl.col("taxonomy_id") == 7955)
)

togoid_convert_join.write_csv(
    "../Data/Wikipathways/WP497-uniprot_taxonomy_converted.tsv",
    separator="\t",
    include_header=True
)

display(togoid_convert_join)
```
