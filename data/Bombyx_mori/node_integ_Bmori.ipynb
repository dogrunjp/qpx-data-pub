```python
# wiki_pathwaysのデータノードファイルと、B.mori遺伝子対応表(DOI: 10.3390/insects13070586)を読み込み
import pandas as pd
node = pd.read_csv('WP497-datanodes.tsv', sep='\t')
Bmori = (
    pd.read_csv('FF4I-B_mori-protein.tsv', sep='\t')
      .rename(columns={
          'H_sapiens-gsymbol': 'Label'
      })
)
display(Bmori)
```

```python
# 対応表と紐づいた遺伝子リストを抽出
merged = node.merge(Bmori, on='Label', how='left')
display(merged)
merged.to_csv('merged.tsv', sep='\t', index=False)
```

```python
# B_mori-geneid を作成して mergedWP534_with_geneid.tsv を出力
import pandas as pd

merged = pd.read_csv('merged.tsv', sep='	', dtype=str)

def build_bmori_geneid(pid: str) -> str:
    if pid is None:
        return ''
    pid = str(pid).strip()
    if pid == '' or pid.lower() == 'nan':
        return ''
    if pid.startswith('KWMTBOMO'):
        return pid.split('.', 1)[0]
    if pid.startswith('MSTRG'):
        parts = pid.split('.')
        return '.'.join(parts[:3]) if len(parts) >= 3 else pid
    return pid.split('.', 1)[0]

bmori_geneid = merged['B_mori-pid'].apply(build_bmori_geneid)
merged.insert(1, 'B_mori-geneid', bmori_geneid)
merged.to_csv('merged_with_geneid.tsv', sep='	', index=False)
print('Wrote: merged_with_geneid.tsv')
```

```python
# 対応表に無い遺伝子リストを抽出して保存するコード
# 1) データ読み込み
# df = pd.read_csv("/mnt/data/mergedWP534.tsv", sep="\t")

# 2) Type が GeneProduct の行を抽出
merged_gp = merged[merged["Type"] == "GeneProduct"].copy()
# Ensemblの接頭辞 'ensembl:' を削除
merged_gp['Ensembl'] = (
    merged_gp['Ensembl'].fillna('').astype(str).str.replace('^ensembl:', '', regex=True)
)

# 3) B_mori-pid が空欄（NaN または 空文字）の行を抽出
df_non_merged = merged_gp[merged_gp["B_mori-pid"].isna() | (merged_gp["B_mori-pid"].astype(str).str.strip() == "")]

# 4) TSV として書き出し
df_non_merged.to_csv('non_merged.tsv', sep="\t", index=False)
```

# 対応表になかった遺伝子に対して以下の作業を行う
- Ensembl bioMartでEnsembl Gene IDをクエリとしてUniProt IDを取得(MANE Selectでフィルタ)
- Uniprot IDをもとにUniProtKBでヒトの対応するアミノ酸配列を取得
- TSAでカイコの全トランスクリプト質配列データ(ICPK00000000.1)を取得
- blastnで確認(クエリ：ヒトアミノ酸、データベース：カイコ核酸)

```python
# non_merged.tsv に UniProtKB/Swiss-Prot ID を付与
import pandas as pd

non_merged = pd.read_csv("non_merged.tsv", sep="\t", dtype=str)
mart = pd.read_csv("mart_export.txt", sep="\t", dtype=str)

# Ensembl == Gene stable ID をキーに付与
mart_sub = mart[["Gene stable ID", "UniProtKB/Swiss-Prot ID"]].drop_duplicates()
non_merged = non_merged.merge(mart_sub, left_on="Ensembl", right_on="Gene stable ID", how="left")
non_merged_filter = non_merged[["Label", "B_mori-geneid", "Type", "Identifier", "Comment", "Ensembl", "NCBI gene", "HGNC", "UniProt", "UniProtKB/Swiss-Prot ID"]]

non_merged_filter.to_csv("non_merged_uniprot.tsv", sep="\t", index=False)
print("Updated: non_merged_uniprot.tsv")
```

```python
# non_merged.tsv（対応無し遺伝子リスト） に、best_by_query_withBmGene_withUniProt.tsv（Blastの結果） を結合するコード
import pandas as pd

# 入力ファイル
f_left = "non_merged_uniprot.tsv"                       # 基準（左）
f_map  = "best_by_query_withBmGene_withUniProt.tsv"   # 対応表（右）

# 読み込み
df_left = pd.read_csv(f_left, sep="\t", dtype=str)
df_map  = pd.read_csv(f_map,  sep="\t", dtype=str)

key = "UniProtKB/Swiss-Prot ID"

# 結合（left join）
merged = df_left.merge(df_map, on=key, how="left")

# 1) 全カラム版を出力
merged.to_csv("non_merged_blast.tsv", sep="\t", index=False)

# 2) 見やすい4列版を出力
# out_4cols = merged[["Label", "B_mori_gene", key, "evalue"]]
# out_4cols.to_csv("non_merged_WP534_blast_4cols.tsv", sep="\t", index=False)

print("Wrote: non_merged_blast.tsv")
# print("Wrote: non_merged_WP534_blast_4cols.tsv")
```

```python
# B_mori-geneid と B_mori_gene の重複を確認して Already_merged を付与
import pandas as pd

merged_geneid = pd.read_csv("merged_with_geneid.tsv", sep="	", dtype=str)
blast_all = pd.read_csv("non_merged_blast.tsv", sep="	", dtype=str)

merged_geneid["B_mori-geneid"] = merged_geneid["B_mori-geneid"].fillna("").str.strip()
blast_all["B_mori_gene"] = blast_all["B_mori_gene"].fillna("").str.strip()

dup_geneids = sorted(set(merged_geneid["B_mori-geneid"]) & set(blast_all["B_mori_gene"]))

if dup_geneids:
    dup_map = (
        merged_geneid[merged_geneid["B_mori-geneid"].isin(dup_geneids)]
        .groupby("B_mori-geneid")["Label"]
        .apply(lambda s: ";".join(sorted({v for v in s.dropna().astype(str) if v.strip()})))
    )
    blast_all["Already_merged"] = blast_all["B_mori_gene"].map(dup_map).fillna("")
    blast_all.to_csv("non_merged_blast_dup.tsv", sep="	", index=False)
    print(f"Updated Already_merged for {blast_all["Already_merged"].astype(bool).sum()} rows")
else:
    if "Already_merged" not in blast_all.columns:
        blast_all["Already_merged"] = ""
    blast_all.to_csv("non_merged_blast_dup.tsv", sep="	", index=False)
    print("No duplicates between B_mori-geneid and B_mori_gene")
```
